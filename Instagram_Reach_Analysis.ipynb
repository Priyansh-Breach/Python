```python
# Importing the libraries.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor
```

```python
# Importing the data

data = pd.read_csv('Instagram.csv', encoding='latin1')
print(data.head)
```

```python
data.isnull().sum()  # Checking for null values
```

```python
data = data.dropna() # Removing null values
```

```python
data.info() # Data types and count.
```

```python
plt.figure(figsize=(10,8))
plt.style.use('fivethirtyeight')
plt.title('Distribution of impressions from home')
sns.distplot(data['From Home'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title('Distribution of impressions from hashtags')
sns.distplot(data['From Hashtags'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title('Distribution of impressions from explorem')
sns.distplot(data['From Explore'])
plt.show()
```

```python
home = data['From Home'].sum()
hashtags = data['From Hashtags'].sum()
explore = data['From Explore'].sum()
other = data['From Other'].sum()

labels = ['From Home', 'From Hashtags', 'From Explore', 'From Other']
values = [home, hashtags, explore, other]

fig = px.pie(data, values = values, labels = labels, title = 'Impressions on Instagram Posts From Various Sources', hole=0.5)
fig.show()
```

```python
text = " ".join(i for i in data.Caption)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.style.use('classic')
plt.figure( figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = " ".join(i for i in data.Hashtags)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.style.use('classic')
plt.figure( figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
conversion_rate = (data["Follows"].sum() / data["Profile Visits"].sum()) * 100
print(conversion_rate)
```

```python
figure = px.scatter(data_frame = data, x="Profile Visits",
                    y="Follows", size="Follows", trendline="ols", 
                    title = "Relationship Between Profile Visits and Followers Gained")
figure.show()
```

```python
x = np.array(data[['Likes', 'Saves', 'Comments', 'Shares', 
                   'Profile Visits', 'Follows']])
y = np.array(data["Impressions"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = PassiveAggressiveRegressor()
model.fit(xtrain, ytrain)
model.score(xtest, ytest)
```

```python
features = np.array([[282.0, 233.0, 4.0, 9.0, 165.0, 54.0]])
model.predict(features)
```
